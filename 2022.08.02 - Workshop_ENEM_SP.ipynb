```python
def linhas_duplicadas(data_frame):
    """Recebe um DataFrame e retorna a quantidade de linhas duplicadas."""
    contagem = data_frame.duplicated().sum()
    return contagem

def colunas_constantes(data_frame):
    """Recebe um DataFrame e retorna uma lista com colunas constantes."""
    return data_frame.columns[data_frame.nunique() == 1].tolist()

def quantidade_missing_values(data_frame):
    """Retorna a quantidade de missing values por coluna."""
    return data_frame.isna().sum()

def pct_missing_values(data_frame):
    """Retorna o percetual de missing values."""
    total_linhas = data_frame.shape[0]
    return np.multiply(data_frame.isna().sum()/total_linhas, 100).round(2)
```

# Função de limpeza da base de dados do ENEM

```python
def limpar_dados_enem(df_case_enem):
    lista_colunas_constantes = colunas_constantes(df_case_enem)
    colunas_nan = df_case_enem.columns[pct_missing_values(df_case_enem) >= 60].tolist()
    tx_colunas_para_remover = ['TX_RESPOSTAS_CN',
    'TX_RESPOSTAS_CH',
    'TX_RESPOSTAS_LC', 
    'TX_RESPOSTAS_MT',
    'TX_GABARITO_CN',
    'TX_GABARITO_CH',
    'TX_GABARITO_LC',
    'TX_GABARITO_MT']
    colunas_para_remover = lista_colunas_constantes+colunas_nan+tx_colunas_para_remover
    df_case_enem_clean = df_case_enem.drop(columns = colunas_para_remover)
    df_case_enem_clean = df_case_enem_clean.dropna()
    return df_case_enem_clean
```

# Base de dados do enem após limpeza

```python
df_case_enem_clean = limpar_dados_enem(df_case_enem)
```
